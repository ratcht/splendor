```python
import sys
from pathlib import Path

sys.path.append(str(Path.cwd().parent))

import numpy as np
from tqdm.auto import tqdm

from env.grid_world import Environment
from grid_world import Parameters, select_action, update
```

```python
env = Environment(dim=(8, 8))

# init q table
params = Parameters(alpha=0.1, epsilon=0.1, gamma=0.99)
num_episodes = 1000
max_steps = 5000
```

```python
print("starting training")

q_table = np.zeros((*env.dim, 4))

rewards = []

pbar_ep = tqdm(range(num_episodes), desc="Episodes")
for ep in pbar_ep:
  env.reset()
  total_reward = 0

  action = select_action(q_table, env.pos, params.epsilon)

  for step in range(max_steps):  # loop until terminal state
    reward, done = env.step(action)

    next_action = select_action(q_table, env.pos, params.epsilon)

    update(q_table, env.prev_pos, env.pos, action, next_action, reward, done, params)

    action = next_action

    total_reward += reward

    if done:
      break
    
  rewards.append(total_reward)
  pbar_ep.set_postfix(steps=step + 1, reward=total_reward, eps=params.epsilon)
```

```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Scatter(
  y=rewards,
  mode='lines',
  name='Raw',
  line=dict(color='steelblue')
))

fig.update_layout(
  title='Training Rewards',
  xaxis_title='Episode',
  yaxis_title='Total Reward',
)

fig.show()
```

```python
# dump the curve values for analysis
print(f"episodes: {len(rewards)}")
print(f"min={min(rewards)}  max={max(rewards)}  mean={np.mean(rewards):.2f}")
print(rewards)
```
